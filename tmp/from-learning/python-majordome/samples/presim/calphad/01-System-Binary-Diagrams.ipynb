## $Al_2O_3-CaO$

```python
%%time

x = [
    0.000,  # 00 CaO
    0.100,  # 01 CaO-C3A1
    0.250,  # 02 C3A1
    0.294,  # 03 Peritectic
    0.360,  # 04 Eutectic
    0.500,  # 05 C1A1
    0.600,  # 06 C1A1-C1A2
    2/3,    # 07 C1A2
    0.700,  # 08 C1A2-C1A6
    0.800,  # 09 C1A2-C1A6
    6/7,    # 10 C1A6
    1.000   # 11 Al2O3
]

rng = (1200.0, 3300.0)
T = np.linspace(*rng, 200)

with Capturing():
    sim = SystemAl2O3CaO(max_iter=10_000)

frames, fig = scan_binary(sim, T, x, "$\\mathrm{Al_2O_3}$", xlim=rng, post=False)
```
